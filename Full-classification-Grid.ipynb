```python
import pandas as pd
```

```python
def DTclassifier_grid(independent,dependent):
    from sklearn.model_selection import GridSearchCV
    from sklearn.tree import DecisionTreeClassifier
    param_grid= {
        'criterion':['gini', 'entropy'],
        'splitter':['best', 'random'],
        'max_features':['sqrt','log2']}
    grid=GridSearchCV(DecisionTreeClassifier(), param_grid, refit=True, verbose=3, n_jobs=1)
    grid.fit(independent, dependent)
    return grid
```

```python
def RFclassifier_grid(independent,dependent):
    from sklearn.model_selection import GridSearchCV
    from sklearn.ensemble import RandomForestClassifier
    param_grid= {
        'criterion':['gini', 'entropy'],
        'n_estimators':[50,100,150],
        'max_features':['sqrt','log2']}
    grid=GridSearchCV(RandomForestClassifier(), param_grid, refit=True, verbose=3, n_jobs=1)
    grid.fit(independent, dependent)
    return grid
```

```python
def Logistic_grid(independent,dependent):
    from sklearn.model_selection import GridSearchCV
    from sklearn.linear_model import LogisticRegression
    param_grid= {
        'penalty':['l1', 'l2'],
        'C':[0.01, 0.1, 1, 10],
        'solver':['liblinear', 'saga'],
        'max_iter':[200,500,1000]}
    grid=GridSearchCV(LogisticRegression(), param_grid, refit=True, verbose=3, n_jobs=1)
    grid.fit(independent, dependent)
    return grid
```

```python
def NB(independent,dependent):
    from sklearn.model_selection import GridSearchCV
    from sklearn.naive_bayes import GaussianNB
    param_grid= {
        'var_smoothing':[1e-9, 1e-8, 1e-7, 1e-6]}
    grid=GridSearchCV(GaussianNB(), param_grid, refit=True, verbose=3, n_jobs=1)
    grid.fit(independent, dependent)
    return grid
```

```python
def SVM(independent,dependent):
    from sklearn.model_selection import GridSearchCV
    from sklearn.svm import SVC
    param_grid= {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf'],
        'gamma': ['scale', 'auto']
        }
    grid=GridSearchCV(SVC(), param_grid, refit=True, verbose=3, n_jobs=1)
    grid.fit(independent, dependent)
    return grid
```

```python
def KNN(independent,dependent):
    from sklearn.model_selection import GridSearchCV
    from sklearn.neighbors import KNeighborsClassifier
    param_grid= {
        'n_neighbors': [3, 5, 7, 9],
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan', 'minkowski']
    }
    grid=GridSearchCV(KNeighborsClassifier(), param_grid, refit=True, verbose=3, n_jobs=1)
    grid.fit(independent,dependent)
    return grid
```

```python
dataset=pd.read_csv("CKD.csv")
dataset=pd.get_dummies(dataset,drop_first=True)

independent=dataset.drop('classification_yes',axis=1)
dependent=dataset['classification_yes']

#grid = DTclassifier_grid(independent,dependent)
#grid = RFclassifier_grid(independent,dependent)
#grid = Logistic_grid(independent,dependent)
#grid = NB(independent,dependent)
#grid = SVM(independent,dependent)
grid = KNN(independent,dependent)

#classifier = DTree(x_train,y_train)
print(f'Best params identified by Grid {grid.best_params_}')
print(f'Results values identified by Grid {grid.cv_results_}')
```
